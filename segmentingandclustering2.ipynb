### Scraping Wikipedia Page Using Pandas

```python
html = wp.page("List_of_postal_codes_of_Canada:_M").html().encode("UTF-8")
df = pd.read_html(html)[0]
df.to_csv('beautifulsoup_pandas.csv',header=0,index=False)
df=df[df.Borough != 'Not assigned']
df=df.astype('str')
```

###  Grouping Neighbourhoods with same Postcode separated by ','

```python
df1 = df.groupby('Postcode')['Neighbourhood'].apply(', '.join).reset_index()
```

### Increasing the numner of rows being displayed  

```python
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)
```

### Removing Duplicates in Postcodes

```python
df2=df
df2.drop_duplicates(subset='Postcode', inplace=True)
```

### Sorting and Setting Post code as the index

```python
df2.sort_values('Postcode', axis=0, ascending=True, inplace=True)
```

```python
df3=df2
```

```python
df3.set_index('Postcode')
```

```python
df4=df1
```

### Setting Postcode as the index for the neighbourhood list grouped based on Postcodes 

```python
df4.set_index('Postcode')
```

### Merging both the data frames and final formatting

```python
df5=pd.merge(df3,df4,left_on='Postcode', right_on='Postcode',how='inner' )
```

```python
df5=df5.drop(columns='Neighbourhood_x')
```

```python
df5.rename(columns={'Postcode':'PostalCode', 'Neighbourhood_y':'Neighbourhood'}, inplace=True)
```
